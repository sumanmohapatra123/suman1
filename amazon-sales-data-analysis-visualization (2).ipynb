```python
df["promotion-ids"]
```

```python
df['ship-city'].fillna('unknown', inplace = True)
df['ship-state'].fillna('unknown', inplace = True)
df['ship-postal-code'].fillna('unknown', inplace = True)
```

## Data Visualization

## Categorical Analysis

 ## 1-Sales by Category

```python
plt.figure(figsize=(10,5))
cat_sales = df.groupby("Category")["Amount"].sum().sort_values(ascending=False)

sns.barplot(
    x=cat_sales.index,
    y=cat_sales.values,
    palette="Set2"
)

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

## 2-Quantity Sold by Category

```python
plt.figure(figsize=(10,5))
cat_qty = df.groupby("Category")["Qty"].sum()

sns.barplot(
    x=cat_qty.index,
    y=cat_qty.values,
    palette="pastel"
)

plt.title("Quantity Sold by Category")
plt.xticks(rotation=45)
plt.show()

```

## Sales by Size

```python
plt.figure(figsize=(8,5))
size_sales = df.groupby("Size")["Amount"].sum()

sns.barplot(
    x=size_sales.index,
    y=size_sales.values,
    palette="coolwarm"
)

plt.title("Sales by Size")
plt.show()
```

## Sales by Fulfilment Method

```python
plt.figure(figsize=(6,5))
full_sales = df.groupby("Fulfilment")["Amount"].sum()

sns.barplot(
    x=full_sales.index,
    y=full_sales.values,
    palette="viridis"
)

plt.title("Sales by Fulfilment Method")
plt.show()
```

## Courier Status Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(
    x="Courier Status",
    data=df,
    palette="Accent"
)

plt.title("Courier Status Distribution")
plt.xticks(rotation=30)
plt.show()

```

## Time Series Analysis

## Sales Trend Over Time

```python
df["Date"] = pd.to_datetime(df["Date"])
daily_sales = df.groupby("Date")["Amount"].sum()

plt.figure(figsize=(12,5))
plt.plot(
    daily_sales.index,
    daily_sales.values,
    marker="o"
)

plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

## Quantity Sold Over Time

```python
daily_qty = df.groupby("Date")["Qty"].sum()

plt.figure(figsize=(12,5))
plt.plot(
    daily_qty.index,
    daily_qty.values,
    marker="o"
)

plt.title("Quantity Sold Over Time")
plt.show()
```

## Distribution Analysis

```python
plt.figure(figsize=(8,5))
sns.histplot(
    df["Amount"],
    kde=True,
    color="Pink"
)

plt.title("Sales Amount Distribution")
plt.show()
```

 ## Quantity Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(
    df["Qty"],
    kde=True,
    color="salmon"
)

plt.title("Quantity Distribution")
plt.show()
```

## Boxplot of Sales Amount

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    y=df["Amount"],
    color="lightgreen"
)

plt.title("Boxplot of Sales Amount")
plt.show()
```

## Boxplot Amount by Category

```python
plt.figure(figsize=(10,5))
sns.boxplot(
    x="Category",
    y="Amount",
    data=df,
    palette="Set2"
)

plt.title("Sales Amount by Category")
plt.xticks(rotation=45)
plt.show()
```

## Top Products Analysis

## Top 10 SKUs by Revenue

```python
top_sku = df.groupby("SKU")["Amount"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
sns.barplot(
    x=top_sku.values,
    y=top_sku.index,
    palette="rocket"
)

plt.title("Top 10 SKUs by Revenue")
plt.xlabel("Total Sales")
plt.show()
```

## Top 10 ASINs by Quantity Sold

```python
top_asin = df.groupby("ASIN")["Qty"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
sns.barplot(
    x=top_asin.values,
    y=top_asin.index,
    palette="mako"
)

plt.title("Top 10 ASINs by Quantity Sold")
plt.show()
```

## B2B vs B2C Sales

```python
b2b_sales = df.groupby("B2B")["Amount"].sum()

plt.figure(figsize=(6,6))
plt.pie(
    b2b_sales,
    labels=["B2C", "B2B"],
    autopct="%1.1f%%",
    startangle=90,
    colors=["#66c2a5", "#fc8d62"]
)

plt.title("B2B vs B2C Sales")
plt.show()
```

## Quantity by B2B

```python
plt.figure(figsize=(6,5))
sns.barplot(
    x="B2B",
    y="Qty",
    data=df,
    estimator=sum,
    palette="Set1"
)

plt.title("Quantity Sold: B2B vs B2C")
plt.show()
```

## Scatter: Quantity vs Amount

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="Qty",
    y="Amount",
    data=df,
    color="purple"
)

plt.title("Quantity vs Amount")
plt.show()
```

## Correlation Heatmap

```python

plt.figure(figsize=(8,5))
sns.heatmap(
    df[["Qty", "Amount", "B2B"]].corr(),
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")
plt.show()
```
